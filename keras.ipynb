```python

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.metrics import MeanAbsoluteError, RootMeanSquaredError
from scikeras.wrappers import KerasRegressor
```

```python
n_features = X.shape[1]

# Define a method to create a model.
def DNN_model(layers,lr_rate):
    DNN = Sequential()
    DNN.add(Input(shape=(n_features,)))
    for i, nodes in enumerate(layers):
        DNN.add(Dense(nodes,kernel_initializer = 'uniform',activation="relu"))
#        if i == 0:
#            DNN.add(Dense(nodes, kernel_initializer = 'glorot_uniform',activation="relu"))
#        else:
#            DNN.add(Dense(nodes,kernel_initializer = 'uniform',activation="relu"))
    DNN.add(Dense(units = 2, kernel_initializer = 'uniform', activation = 'softmax'))
    opt = SGD(learning_rate=lr_rate)
    DNN.compile(
        optimizer = opt,
        loss="mse",
        metrics=[RootMeanSquaredError()]
        )    
    return DNN
```
